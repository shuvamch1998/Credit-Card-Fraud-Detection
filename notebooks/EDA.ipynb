```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/SHUVAM CHATTERJEE/Downloads/Creditcard_Fraud_Detection/data/creditcard.csv')
df.head()
```

```python
df.info()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(data=df, x='Class')
plt.title('Fraud vs Non Fraud Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  
sns.countplot(x='Hour', hue='Class', data=df)
plt.title('Fraud Distribution Across Hours', fontsize=14)
plt.xlabel('Hour of Day')
plt.ylabel('Transaction Count')
plt.legend(title='Class', labels=['Not Fraud (0)', 'Fraud (1)'])
plt.tight_layout()  
plt.show()
```

```python
corr_matrix = df.corr()
fraud_corr = corr_matrix['Class'].sort_values(ascending=False)
print(fraud_corr.head(10))
```

```python
top_features = fraud_corr[1:6].index  
for feature in top_features:
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f'{feature} vs Class')
    plt.show()
```
